```python
!pip install nltk
```

Import nltk library and word_tokenize function from nltk.tokenize

```python
import nltk
from nltk.tokenize import word_tokenize
```

Downloading the 'punkt' tokenizer model 'punkt' is a pre-trained model used by NLTK for tokenizing text into sentences or words.

```python
nltk.download('punkt')
```

Create a variable ‘text’ that contains a paragraph of at least 4 sentences about any topic of your choice.

```python
text = 'A breakfast-room adjoined the drawing-room, I slipped in there. It contained a bookcase: I soon possessed myself of a volume, taking care that it should be one stored with pictures. I mounted into the window-seat: gathering up my feet, I sat cross-legged, like a Turk; and, having drawn the red moreen curtain nearly close, I was shrined in double retirement. Folds of scarlet drapery shut in my view to the right hand; to the left were the clear panes of glass, protecting, but not separating me from the drear November day.'
print(text)
```

*   Use the word_tokenize function to tokenize the text into words.
*   Also I am the lower function to lowered the text in order to not confuse the tokenizer to thinks about same word as different if starts with capital letter.



```python
text_tokenize = word_tokenize(text.lower())
print(text_tokenize)
```

Count the number of tokens in the text.

```python
print(len(text_tokenize))
```

Identify the frequency of each token.

```python
freq_dist = nltk.FreqDist(text_tokenize)
print(freq_dist)
```

Couting each word and its frequency in the text, freq_dist.items() returns each word and its corresponding frequency count.

```python
for word, freq in freq_dist.items():
  print(f"{word}: {freq}")
```

Importing libraries we need to use.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

Load a text data from a file in Google Drive and then read it save it to variable text.

```python
with open('/content/drive/MyDrive/Conversational AI: History, Application, Future/JaneEyre.txt') as file:
  text = file.read()
```

Create a WordCloud object.

```python
wordcloud = WordCloud(
width=800,                     # Width of the word cloud image
height=400,                    # Height of the word cloud image
background_color="white",      # Background color of the word cloud
stopwords=set(STOPWORDS),      # Remove common stopwords (e.g., "the", "and")
min_font_size=10              # Minimum font size for the words
).generate(text)
```

Display the word cloud.

```python
plt.figure(figsize=(8, 8), facecolor=None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

As you see above, words like **now**, **will**, **one**, **said**, **Mr**, **Rochester** appear larger and bolder in the output, indicating they occur more frequently in the text used to generate the word cloud.

Word clouds visually represent word frequency, where the size of each word correlates with its frequency in the text. Commonly used words like **said** and **one** are often prevalent in narrative texts, as character's dialogues and general descriptions tend to include these terms frequently. "Mr. Rochester" is likely prominent because it could be the name of a central character, making it a repeated reference.

Words such as "thought," "know," and "will" might reflect common themes or expressions in the text, particularly if it’s a work of literature where characters reflect on actions, feelings, and plans.

### **Customization**
Experiment with the WordCloud parameters in your Colab notebook to customize your word cloud. Try different:
*   background_color
*   colormap
*   max_words
*   stopwords





```python
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color="green",  # Change background color ('black', "white", "pink")
    colormap="plasma",           # Try different color maps (e.g., "viridis", "plasma", "cividis", "copper")
    max_words=50,              # Limit the number of words displayed
    stopwords=set(STOPWORDS),  # We can also use custom stopwords as well
    min_font_size=10
).generate(text)
```

```python
# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

### **Mask**
Research how to use the mask parameter to create a word cloud in a specific shape. Upload a mask image to your Drive and use it in your code.
